DATA FRAME 1 - bitcoin_articles.csv

Corresponde a una base de datos de artículos informativos sobre bitcoin.
Las columnas más relevantes de este dataset son:

link -> aporta el enlace al artículo informativo correspondiente

title -> presenta la temática del artículo

rank -> informa cuáles son los artículos que tuvieron más alcance


DATA FRAME 2 - BTC_Data_final.csv

Es un data frame con información sobre la cotización de bitcoin desde el año 2014 hasta fin del año 2022.

Las columnas más relevantes de este dataset son:

Date -> aporta la fecha de la fila correspondiente

priceUSD -> es el dato del valor de bitcoin en dolares

marketcap -> informa la capitalización de mercado total de bitcoin a la fecha


DATA FRAME 3 - NFTData.csv

Es un dataset que contiene información sobre el top200 de colecciones de NFTS en la blockchain Ethereum.

Las columnas más relevantes de este dataset son:

Name -> es el nombre de la coleccion de NFT

Market_Cap_USD -> aporta el valor total de la capitalizaciÓn de mercado de la colecciÓn

Floor_Price_USD -> es el precio más bajo al que podemos encontrar un NFT en esta colecciÓn
