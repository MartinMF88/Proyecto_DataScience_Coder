```python
q1 = df['tweets'].quantile(0.2)
q3 = df['tweets'].quantile(0.5)
iqr = q3 - q1
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

outliers = df[(df['tweets'] < lower_limit) | (df['tweets'] > upper_limit)]

df_cleaned = df[(df['tweets'] >= lower_limit) & (df['tweets'] <= upper_limit)]

plt.boxplot(df_cleaned['tweets'])
plt.title('tweets')
plt.show()
```

```python
plt.scatter(df['tweets'], df['priceUSD'], s=10, c='blue')

plt.plot(linewidth=5, alpha=0.60)

plt.title('Relación entre tweets y el precio de Bitcoin')
plt.xlabel('tweets')
plt.ylabel('Precio de Bitcoin (en USD)')
plt.grid(True)
plt.show()
```
